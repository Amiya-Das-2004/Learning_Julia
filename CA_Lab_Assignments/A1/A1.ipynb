```julia
using CalculusWithJulia
using Plots
using LaTeXStrings
using Symbolics
plotly()
```

```julia
gr()

h(x,y) = 200 - (x^2) - (2*(y^2))
h(v) = h(v...)
```

```julia
x = -15:0.1:15
y = -15:0.1:15;
```

```julia
z = [h(xi, yi) for yi in y, xi in x];
```

```julia
D = surface(x, y, z,
    xlabel="x", ylabel="y", zlabel="h(x,y)",
    title=L"3D\ Surface:\ h(x,y) = 200 - x^2 - 2y^2",
    color=:viridis)
```

```julia
contour(x, y, z,
    xlabel="x", ylabel="y",
    title=L"2D Contour Plot of h(x,y)",
    color=:viridis)
```
