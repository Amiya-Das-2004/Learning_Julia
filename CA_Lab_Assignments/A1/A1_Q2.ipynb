```julia
using Plots
using LaTeXStrings
gr()
```

```julia
SFD(x,q) = ((5*L/8)-x)*q
```

```julia
L = 10
x = 0:0.1:L
q = 0.2
```

```julia
y = SFD.(x,q);
```

```julia
p = plot(x, y)
plot!(x,x*0)
plot!(y*0,y)
```

```julia
a = L:-0.1:0
BMD(x,q) = (((3*L/8)-(x/2))*q*x)
y1 = BMD.(a,q);
```

```julia
p = plot(x,y1, 
    legend=false,
    xlabel=L"x", ylabel=L"y",
    grid=true,
    gridcolor=:black,      # color of the grid
    gridalpha=1,         # transparency (0 = invisible, 1 = solid)
    gridstyle=line,        # dotted grid lines
    xminorgrid=true,       # add minor grid lines for finer mesh
    yminorgrid=true,
    minorgridcolor=:black,
    minorgridalpha=0.7,
    minorgridstyle=:dot)
plot!([0, 0], [0, -2], color=:red, linestyle=:dot, lw=2, label="")
plot!(x,0*x)
```
